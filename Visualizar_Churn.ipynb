```python
# Identifico el nombre del archivo subido y lo leo con pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataframe
df = pd.read_csv(list(uploaded.keys())[0])

# Etiquetas descriptivas en las variables categóricas
#df['churn'] = df['churn'].replace({1: 'Perdido', 0: 'Retenido'})
df['credit_card'] = df['credit_card'].replace({1: 'Sí', 0: 'No'})
df['active_member'] = df['active_member'].replace({1: 'Sí', 0: 'No'})
df.head()

# Me aseguro de ver la forma inicial y un vistazo rápido
print("Dimensiones iniciales: ", df.shape)
display(df.head())
```

# **2. Visualización con Plotly Express**

## **Instalar y acutalizar librerías**

```python
# Instalo plotly y kaleido (kaleido me permite exportar PNG de las figuras)
#!pip install plotly kaleido
#!pip install --upgrade plotly

# Llamo librerías
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "colab"  # para mostrar en Colab
```

## **Dispersión agrupada por tamaño y color**

```python
# Gráfico de puntos
fig = px.scatter(df.sample(1000),
                 x='age', y='balance', color='churn',
                 size='products_number', #symbol='churn',
                 title='Comparación entre Edad y Saldo, agrupando por Productos Bancarios y Churn'
                 )
fig.show()
```

## **Mapa de calor con dispersión marginal**

```python
fig = px.density_heatmap(df,
                         x = 'age', y = 'credit_score', z = 'estimated_salary',
                         marginal_x = 'box', marginal_y = 'box',
                         title='Mapa de calor de Score Crediticio y Edad, destacando Salario Estimado'
                         )

fig.show()
```

## **Dendograma**
Dendograma del número de Productos Bancarios por Churn

```python
import plotly.figure_factory as ff
import numpy as np
np.random.seed(2025)

# 20 observaciones, con 5 dimensiones cada una
df2 = df[['churn','products_number']]
data = df2.sample(100)

# Calcular dendograma
fig = ff.create_dendrogram(data)
fig.update_layout(autosize = True)
fig.show()
```

## **Radar**

```python
num_cols = ['credit_score','age','balance','estimated_salary']

# Agrupar por 'country' y calcular el promedio de las columnas numéricas
df_promedio = df.groupby(['country','churn'])[num_cols].mean().reset_index()


# Mostrar el nuevo DataFrame
#display(df_promedio)

# Grafica radar
fig = px.line_polar(df_promedio,
                    r = 'balance',
                    theta = 'country',
                    color = 'churn',
                    line_dash = 'churn',
                    line_close = True)

# Mostrar figura
fig.update_traces(fill = 'toself')
fig.show()
```

## **Embudo**

```python
fig = px.funnel_area(df,
                names = 'country',
                values = 'credit_score',
                title = 'Embudo de Score Crediticio por País'
                )


#fig.update_traces(textfont = {'color': 'white'})
fig.show()
```
